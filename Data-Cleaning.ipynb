```python
import pandas as pd
```

```python
ajio = pd.read_csv('Ajio_Fasion_Clothing.csv', sep=',', encoding='latin1')
ajio
```

```python
ajio['Original Price (in Rs.)'].dtype
```

```python
ajio = ajio.rename(columns={'Product_URL': 'Product_Image', 'Brand': 'brand',
                           'Description' : 'desc', 'Id_Product' : 'id',
                           'URL_image' : 'image', 'Category_by_gender' : 'gender',
                           'Discount Price (in Rs.)' : 'discount', 
                           'Original Price (in Rs.)' : 'original_price',
                           'Color' : 'color'})
```

```python
ajio['discount'] = pd.to_numeric(ajio['discount'].str.replace(',', ''), errors='coerce').fillna(0)
```

```python
ajio['original_price'] = pd.to_numeric(ajio['original_price'].str.replace(',', ''), errors='coerce').fillna(0)
```

```python
print(ajio.columns.tolist())
```

```python
ajio = ajio.drop(columns=['Product_Image','image'], errors='ignore')
```

```python
print(ajio.columns.tolist())
```

```python
ajio = ajio.drop(columns = "")
```

```python
ajio.to_csv('ajio_updated.csv')
```
